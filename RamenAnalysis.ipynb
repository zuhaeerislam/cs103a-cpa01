## Introduction
This dataset holds several ratings for ramen products and records the brand, name, style, country, and average stars of each product. This dataset can be found at https://www.kaggle.com/residentmario/ramen-ratings. 

**Questions**

1. which brand has the highest average ramen rating?
2. which country has the highest average ramen rating?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ramen-ratings.csv")
df
```

```python
df.describe()
```

```python
list(df.columns)
```

```python
list(df.index)
```

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors = 'coerce')
stats = pd.pivot_table(data=df, 
                        values =['Stars'], 
                        index=['Style'], 
                        aggfunc='mean')
print(stats)
stats.plot()
plt.title("Ramen Rating by Type")
plt.ylabel("Avg. Rating")
```

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors = 'coerce')
branddf = df.groupby(['Brand']).mean()
branddf = branddf.sort_values('Stars', ascending = False)
print(branddf.loc[branddf['Stars'] == 5.0] )
```

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors = 'coerce')
styledf = df.groupby(['Style']).mean()
styledf = styledf.sort_values('Stars', ascending = False)
print(styledf['Stars'])
```

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors = 'coerce')
countrydf = df.groupby(['Country']).mean()
countrydf = countrydf.sort_values('Stars', ascending = False)
print(countrydf['Stars'])
```
